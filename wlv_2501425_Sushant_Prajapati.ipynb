```python
import matplotlib.pyplot as plt

#outlier Detection
def outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

hdi_south_asia['HDI_Outliers'] = outliers(hdi_south_asia['hdi'])
hdi_south_asia['GNI_Outliers'] = outliers(hdi_south_asia['gross_inc_percap'])

#plotting graph
plt.figure()

normal = hdi_south_asia[~(hdi_south_asia['HDI_Outliers'] | hdi_south_asia['GNI_Outliers'])]
outliers = hdi_south_asia[hdi_south_asia['HDI_Outliers'] | hdi_south_asia['GNI_Outliers']]

plt.scatter(normal['gross_inc_percap'], normal['hdi'], label='Normal')
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'], label='Outliers')

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("GNI vs HDI (Outliers Highlighted)")
plt.legend()
plt.show()
```

Discuss why the identified countries stand out as outliers.

=> Countries like Maldives stand out due to significantly higher GNI compared to regional peers, whereas Afghanistan appears as a low-HDI outlier due to the prolonged conflict and limited human development.

Descriptive Statistics

```python
# Middle East
mean_me = df_me['hdi'].mean()
std_me = df_me['hdi'].std()

# South Asia
mean_sa = df_sa['hdi'].mean()
std_sa = df_sa['hdi'].std()

print(f"Middle East HDI: mean = {mean_me:.3f}, std = {std_me:.3f}")
print(f"South Asia HDI: mean = {mean_sa:.3f}, std = {std_sa:.3f}")
```

Middle Eastern countries dominate the top HDI ranks, while South Asian countries are more prominent in lower rankings.

```python
#Metric Comperisons

metrics = ['gender_development', 'life_expectancy', 'gross_inc_percap']

for metric in metrics:
    plt.figure(figsize=(10,6))
    # Average by region
    avg_values = pd.DataFrame({
        'Middle East': df_me[metric].mean(),
        'South Asia': df_sa[metric].mean()
    }, index=[metric])
    avg_values.T.plot(kind='bar')
    plt.title(f"Average {metric} Comparison")
    plt.ylabel(metric)
    plt.show()

```

Identify which metric shows the greatest disparity between regions.

```python
#HDI Disparity
#HDI Range
range_me = df_me['hdi'].max() - df_me['hdi'].min()
range_sa = df_sa['hdi'].max() - df_sa['hdi'].min()

#cofficient of varitian
cv_me = std_me / mean_me
cv_sa = std_sa / mean_sa
print(f"Middle East HDI Range = {range_me:.3f}, CV = {cv_me:.3f}")
print(f"South Asia HDI Range = {range_sa:.3f}, CV = {cv_sa:.3f}")
```

 Correlations Analysis

```python
for df_region, name in [(df_me, 'Middle East'), (df_sa, 'South Asia')]:

    # Correlations Analysis
    corr_gender = df_region['hdi'].corr(df_region['gender_development'])
    corr_life = df_region['hdi'].corr(df_region['life_expectancy'])

    print(f"{name} - HDI vs gender_development: {corr_gender:.3f}")
    print(f"{name} - HDI vs life_expectancy: {corr_life:.3f}")

    # Scatter plots with trendlines
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.regplot(data=df_region, x='gender_development', y='hdi')
    plt.title(f"{name}: HDI vs Gender Development")
    plt.xlabel('Gender Development Index')
    plt.ylabel('HDI')

    plt.subplot(1, 2, 2)
    sns.regplot(data=df_region, x='life_expectancy', y='hdi')
    plt.title(f"{name}: HDI vs Life Expectancy")
    plt.xlabel('Life Expectancy Index')
    plt.ylabel('HDI')

    plt.tight_layout()
    plt.show()
```

Outlier Detection:

```python
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df[(df[col] < lower) | (df[col] > upper)]


for df_region, name in [(df_me, 'Middle East'), (df_sa, 'South Asia')]:
    for col in ['hdi', 'gross_inc_percap']:

        outliers = detect_outliers(df_region, col)

        print(f"{name} - Outliers in {col}:")
        display(outliers[['country', col]])

        # Scatter plot highlighting outliers
        plt.figure(figsize=(7, 4))

        plt.scatter(df_region.index, df_region[col], label='Normal')
        plt.scatter(outliers.index, outliers[col], label='Outlier')

        plt.xlabel('Country Index')
        plt.ylabel(col)
        plt.title(f"{name} - {col} Outlier Detection (IQR Method)")
        plt.legend()
        plt.show()
```

Discuss the significance of these outliers.

=> Countries identified as outlier significantly deviate from the regional norm in terms of HDI or GNI per capita
High-GNI outliers often represent oil-rich economies, while low-HDI outliers indicate nations affected by conflict, policies instability, or limited access to social services
